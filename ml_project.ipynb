This is done to reduce the target classes (bucketize them into 4 classes) based on their performance.

```python
def red_classes(dfa):
  for i in range(0,dfa.shape[0]):
    if(dfa.iloc[i]<=5):
      dfa.iloc[i] = 0 #low performance
    elif(dfa.iloc[i]<=10):
      dfa.iloc[i] = 1#below average
    elif(dfa.iloc[i]<=15):
      dfa.iloc[i] = 2 #above average
    else:
      dfa.iloc[i]= 3  #high performance
  return dfa
```

Running 100 complete epochs on the data , so that we can plot the loss_curves

```python
from sklearn.metrics import log_loss
import numpy as np
def run_epoch(epoch,x_val,y_val):
  for i in range(epoch):
    clf.partial_fit(x_val,y_val, np.unique(y_val))
    # loss_on_train = clf.loss_
    loss_t.append(clf.loss_)
    pred = clf.predict_proba(x_test)
    loss_v.append(log_loss(y_test, pred))
```

```python
for i in range(epoch):
  clf.partial_fit(x_train,y_train, np.unique(y_train))
  # loss_on_train = clf.loss_
  loss_t.append(clf.loss_)
  pred = clf.predict_proba(x_test)
  loss_v.append(log_loss(y_test, pred))
```

Plotting code for graph

```python
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
#features_scaled,target_scaled
svm = SVR(C=0.25)
random_forest = RandomForestRegressor(max_depth=3)
def ml_algorithm(estimate):
  train_sizes, train_scores, test_scores = learning_curve(
      estimator=estimate,
      X=features_scaled,
      y=target_scaled,
      cv=5,
      train_sizes = [1,30,50,90,100,120,180,240]
  )

  train_mean = train_scores.mean(axis=1)
  test_mean = test_scores.mean(axis=1)

  plt.subplots(figsize=(10,8))
  plt.plot(train_sizes, train_mean, label="train")
  plt.plot(train_sizes, test_mean, label="validation")

  plt.title("Learning Curve")
  plt.xlabel("Training Set Size")
  plt.ylabel("RMSE")
  plt.legend(loc="best")

  plt.show()
```

```python
ml_algorithm(random_forest)
```

```python
ml_algorithm(svm)
```

Performing GridSearchCV on RandomForest Classifier Hyper-parameters to find out the best parameters for classification

```python
from sklearn.ensemble import RandomForestClassifier
n_estimators = [20,30]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [2,4]
# Minimum number of samples required to split a node
min_samples_split = [2, 5]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2]

param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
print(param_grid)

rf_Model = RandomForestClassifier()
from sklearn.model_selection import GridSearchCV
rf_Grid = GridSearchCV(estimator = rf_Model, param_grid = param_grid, cv = 3, verbose=2, n_jobs = 4)

rf_Grid.fit(x_train, y_train)
```
